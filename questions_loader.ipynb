```python
import os
from flask import Flask
import json
from app import Question


THIS_ROOT = os.getcwd()  # sets notebook directory as root

app = Flask(__name__, root_path=THIS_ROOT)
app.config['SECRET_KEY'] = "something-secret"
app.config['SQLALCHEMY_DATABASE_URI'] = "sqlite:///quiz.db"
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False






DATA_DIR = 'data'

FILES = [
    ("Python", "Easy", "python_easy.json"),
    # ("Python", "Medium", "python_medium.json"),
    # ("Statistics", "Easy", "stats_easy.json"),
    # Add more (subject, level, filename) here
]

def load_questions():
    with app.app_context():
        for subject, level, filename in FILES:
            path = os.path.join(DATA_DIR, filename)
            if not os.path.exists(path):
                print(f"File not found: {path}")
                continue

            with open(path, 'r', encoding='utf-8') as f:
                questions = json.load(f)

            for q in questions:
                # avoid duplicate (subject, level, number)
                existing = Question.query.filter_by(
                    subject=subject,
                    level=level,
                    number=q["number"]
                ).first()

                if existing:
                    continue

                question = Question(
                    subject=subject,
                    level=level,
                    number=q["number"],
                    question_text=q["question_text"],
                    option_a=q["option_a"],
                    option_b=q["option_b"],
                    option_c=q["option_c"],
                    option_d=q["option_d"],
                    correct_option=q["correct_option"],
                    explanation=q.get("explanation", "")
                )
                db.session.add(question)

            db.session.commit()
            print(f"Loaded questions for {subject} - {level} from {filename}")

if __name__ == "__main__":
    load_questions()
```
