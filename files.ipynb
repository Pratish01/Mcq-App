```python
import PyPDF2
import json
import re

pdf_file = r"C:\Users\hp\Downloads\Python Snippet Question Bank.pdf"
output_json = "data/python_easy.json"

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text


def parse_mcqs(text):
    # Regex pattern for your exact format
    pattern = re.compile(
        r"(.*?)\n\s*a\)\s*(.*?)\n\s*b\)\s*(.*?)\n\s*c\)\s*(.*?)\n\s*d\)\s*(.*?)\n\s*Answer:\s*([a-dA-D])",
        re.DOTALL
    )

    questions = []
    q_number = 1

    for match in pattern.finditer(text):
        question_text = match.group(1).strip()
        option_a = match.group(2).strip()
        option_b = match.group(3).strip()
        option_c = match.group(4).strip()
        option_d = match.group(5).strip()
        correct = match.group(6).upper()

        questions.append({
            "number": q_number,
            "question_text": question_text,
            "option_a": option_a,
            "option_b": option_b,
            "option_c": option_c,
            "option_d": option_d,
            "correct_option": {"A": "A", "B": "B", "C": "C", "D": "D"}[correct],
            "explanation": ""
        })

        q_number += 1

    return questions


text = extract_text_from_pdf(pdf_file)
mcqs = parse_mcqs(text)

with open(output_json, "w", encoding="utf-8") as f:
    json.dump(mcqs, f, indent=4)

print(f"Extracted {len(mcqs)} questions into {output_json}")
```
